```python
import pandas as pd
import numpy as np

import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import f1_score

import matplotlib.pyplot as plt
```

```python
#reading the file

df = pd.read_csv("data.csv")
```

```python
df
```

```python
#checking if balanced data

df.info()
```

```python
#taking the content column from the data frame

contents = df['CONTENT']
```

```python
contents
```

```python
#cleaning the dataset; converting to lower; splits into list; removal of stopwords; and Stemming
ps = PorterStemmer()
corpus = []

for i in range(0, len(contents)):
    review = re.sub('[^a-zA-Z]', ' ', contents[i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#convert the corpus to vectors; prepping the data

tfidfcv = TfidfVectorizer()
X = tfidfcv.fit_transform(corpus).toarray()
```

```python
#independent variable
X
```

```python
#dependent variable
y=df['CLASS']
```

```python
y
```

```python
# Train Test Split - split the independent and dependent variables from 80-20 split. Test size is 20%. 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

# 1. Naive Bayes Multinomial Model 

```python
# Training model using Naive bayes classifier

from sklearn.naive_bayes import MultinomialNB

NB_model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred=NB_model.predict(X_test)
```

```python
#confusion matrix - to determine the positive and negative results of the model. 
#Comparing the y_test labels (20% of the original dataset messages) Then, compare the prediction made by the model in y_pred

from sklearn.metrics import confusion_matrix
confusion_mat = confusion_matrix(y_test,y_pred)
```

```python
confusion_mat
```

```python
accuracy = f1_score(y_test,y_pred)
accuracy
```

# 2. Passive Aggressive Classifier Algorithm

```python
from sklearn.linear_model import PassiveAggressiveClassifier

linear_clf = PassiveAggressiveClassifier()
```

```python
linear_clf.fit(X_train,y_train)
```

```python
y_pred=linear_clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_mat = confusion_matrix(y_test,y_pred)
```

```python
confusion_mat
```

```python
accuracy = f1_score(y_test,y_pred)
accuracy
```

# 3. Building Own Model using LSTM

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=2000
sent_length=20
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```python
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## Creating model
embedding_vector_features=40

model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(200))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

print(model.summary())
```

```python
#checking data shapes
len(embedded_docs),y.shape
```

```python
#convert to numpy
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
### Finally Training
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
pred_test = model.predict(X_test)
```

```python
#Evaluation on Training Data

for item in history.history:
    print("Final", item, ":", history.history[item][-1])
```
