```python
import pandas as pd

import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

from sklearn.feature_extraction.text import TfidfVectorizer

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#reading the file

df = pd.read_csv("data.csv")
```

```python
df
```

```python
#checking if balanced data

df.info()
```

```python
#taking the content column from the data frame

contents = df['CONTENT']
```

```python
contents
```

```python
#cleaning the dataset; converting to lower; splits into list; removal of stopwords; and Stemming
ps = PorterStemmer()
corpus = []

for i in range(0, len(contents)):
    review = re.sub('[^a-zA-Z]', ' ', contents[i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#convert the corpus to vectors; prepping the data

tfidfcv = TfidfVectorizer()
X = tfidfcv.fit_transform(corpus).toarray()
```

```python
X
```

```python
y=df['CLASS']
```

# K MEANS Clustering

```python
# Train Test Split - split the independent and dependent variables from 80-20 split. Test size is 20%. 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

# K=2

```python
true_k = 2
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=200, n_init=1)
model.fit(X)
```

```python
#clustering

print("Top terms per cluster:")
centroids = model.cluster_centers_.argsort()[:, ::-1]
words = tfidfcv.get_feature_names()


for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in centroids[i, :10]:
        print(' %s' % words[ind]),
    print
```

# K=3

```python
true_k = 3
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=200, n_init=1)
model.fit(X)
```

```python
#clustering

print("Top terms per cluster:")
centroids = model.cluster_centers_.argsort()[:, ::-1]
words = tfidfcv.get_feature_names()


for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in centroids[i, :10]:
        print(' %s' % words[ind]),
    print
```
